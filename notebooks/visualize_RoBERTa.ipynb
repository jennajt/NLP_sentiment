## Get processed RoBERTa Data

```python
data = pd.read_csv("/content/drive/MyDrive/data/roberta datasets/ukraine_roberta_processed.csv")
tweets_df = pd.DataFrame(data)
tweets_df
```

## Play with your data

```python
#group it by day and get the means of predictions and likes and total likes

group_by_day_df = pd.DataFrame()

mean_by_day_df = tweets_df.groupby("only_date", as_index=False).mean()
sum_by_day_df = tweets_df.groupby("only_date", as_index=False).sum()

group_by_day_df["date"] = mean_by_day_df["only_date"]

group_by_day_df["joy_means"] = mean_by_day_df["joy"]
group_by_day_df["optimism_means"] = mean_by_day_df["optimism"]
group_by_day_df["anger_means"] = mean_by_day_df["anger"]
group_by_day_df["sadness_means"] = mean_by_day_df["sadness"]
group_by_day_df["positive_means"] = mean_by_day_df["positive"]
group_by_day_df["neutral_means"] = mean_by_day_df["neutral"]
group_by_day_df["negative_means"] = mean_by_day_df["negative"]
group_by_day_df["like_means"] = mean_by_day_df["Likecount"]
group_by_day_df["like_totals"] = sum_by_day_df["Likecount"]

group_by_day_df
```
