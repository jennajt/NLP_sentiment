```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
# create list to append tweet data to
tweets_list = []
```

```python
# params
keywords = ''
```

```python
%%time
# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper(f'{keywords} from:TheSun OR from:DailyMailUK OR from:MailOnline OR from:Daily_Express since:2008-07-28 until:2009-07-28').get_items()):
    tweets_list.append([tweet.date, tweet.id, tweet.content, tweet.user.username, tweet.likeCount])
tweets_list
```

```python
len(tweets_list)
```

```python
# Creating a dataframe from the tweets list above
tweets_df = pd.DataFrame(tweets_list, columns=['Datetime', 'Tweet Id', 'Text', 'Username','Like Count'])
tweets_df.head()
```

```python
tweets_df.tail()
```
